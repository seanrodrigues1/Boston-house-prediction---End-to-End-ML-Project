```python
df.shape
```

```python
# separating independent and dependent features


y=df["House_Price"]

x=df.drop(["House_Price"],axis=1)
```

```python
#train test split


from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#scaling values

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

# TRAINING MACHINE LEARNING MODELS

```python
#LINEAR REGRESSION
```

```python
from sklearn.linear_model import LinearRegression

regression=LinearRegression()

regression.fit(x_train,y_train)

```

```python
#REGRESSION COEFFICIENTS

regression.coef_
```

```python
# TRAINING SET ACCURACY

from sklearn.metrics import mean_absolute_error

y_pred=regression.predict(x_train)


score=mean_absolute_error(y_train,y_pred)

print("Mean Absolute Error:",score)
```

```python
#TESTING SET ACCURACY

y_pred=regression.predict(x_test)

score=mean_absolute_error(y_test,y_pred)

print("Mean Absolute Error:",score)
```

```python
#POLYNOMIAL REGRESSION
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(2)                             # SET DEGREE 2

poly_features = poly.fit_transform(x_train)            #WE NEED TO TRANSFORM OUR DEPENDENT FEATURES BEFORE TRAINING

poly_reg_model = LinearRegression()

poly_reg_model.fit(poly_features, y_train)
```

```python
# TRAINING SET ACCURACY

y_pred=poly_reg_model.predict(poly_features)


score=mean_absolute_error(y_train,y_pred)

print("Mean Absolute Error:",score)
```

```python
#TESTING SET ACCURACY

poly_features_test = poly.fit_transform(x_test)

y_pred=poly_reg_model.predict(poly_features_test)

score=mean_absolute_error(y_test,y_pred)

print("Mean Absolute Error:",score)
```

```python
#XGBOOST
```

```python
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor

model=XGBRegressor()

parameters= {"learning_rate" : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,"max_depth": [ 3, 4, 5, 6, 8, 10, 12, 15],"min_child_weight" : [0, 1, 3, 5, 7 ],
 "gamma": [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],"colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7,1 ],"scale_pos_weight":[1,0]}

random_search=RandomizedSearchCV(model,param_distributions=parameters,scoring='neg_mean_absolute_error',cv=5)

random_search.fit(x_train,y_train)

print(random_search.best_params_)
print(random_search.best_score_)
```

```python
xgb_model=XGBRegressor(n_estimators=100,learning_rate=0.15,max_depth=10,min_child_weight=3,gamma=0.1,reg_lambda=1,reg_alpha=0,scale_pos_weight = 1,colsample_bytree= 0.7)

xgb_model.fit(x_train,y_train)
```

```python
# TRAINING SET ACCURACY

y_pred = xgb_model.predict(x_train)


score = mean_absolute_error(y_train,y_pred)

print("Mean Absolute Error:",score)
```

```python
#TESTING SET ACCURACY

y_pred = xgb_model.predict(x_test)

score = mean_absolute_error(y_test,y_pred)

print("Mean Absolute Error:",score)
```
